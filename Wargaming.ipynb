```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
original_df = pd.read_excel('C:/Users/Steven Ma/Desktop/Data_Analyst_Test.xlsx', 'players_list')
print(original_df.columns)
```

```python
original_df = original_df[['battles_cnt', 'win_cnt', 'total_playtime_min', 'first_battle_dt', 'buy_GoldPack4', 'is_payer']]
# original_df = original_df.drop(columns=['first_TankE_date'])
original_df['win_pct'] = original_df['win_cnt'] / original_df['battles_cnt']
is_payer = original_df.loc[original_df['is_payer'] == 1]
non_payer = original_df.loc[original_df['is_payer'] == 0]
buy_goldpack4 = original_df.loc[original_df['buy_GoldPack4'] == 1]
non_goldpack4 = original_df.loc[original_df['buy_GoldPack4'] == 0]
```

```python
from sklearn.tree import DecisionTreeClassifier

X = original_df[['battles_cnt', 'win_cnt', 'total_playtime_min', 'win_pct']]
Y = original_df[['is_payer']]

# define the model
model_cart = DecisionTreeClassifier()
# fit the model
model_cart.fit(X, Y)
# get importance
importance_cart = model_cart.feature_importances_
# summarize feature importance
for i,v in enumerate(importance_cart):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in range(len(importance_cart))], importance_cart)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
X = original_df[['battles_cnt', 'win_cnt', 'total_playtime_min', 'win_pct']]
Y = original_df[['is_payer']]
# define the model
model_rf = RandomForestClassifier()
# fit the model
model_rf.fit(X, Y)
# get importance
importance_rf = model_rf.feature_importances_
# summarize feature importance
for i,v in enumerate(importance_rf):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.bar([x for x in range(len(importance_rf))], importance_rf)
plt.show()
```
